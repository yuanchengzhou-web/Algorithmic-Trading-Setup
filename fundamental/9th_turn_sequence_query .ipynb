```python
import akshare as ak
import yfinance as yf
```

```python
us_stock_current_df = ak.stock_us_spot()
us_stock_current_df
```

```python
stock_num = range(len(us_stock_current_df))
stock_num
```

```python
buy_stocks_days = []
sell_stocks_days = []
buy_stocks_weeks = []
sell_stocks_weeks = []
buy_stocks_months = []
sell_stocks_months = []

for i in stock_num:
    tickers = us_stock_current_df.iloc[i,3]
    selected_ticker = yf.Ticker(tickers)
    hist_days = selected_ticker.history(period="1mo")
    hist_weeks = selected_ticker.history(period='1y', interval='1wk')
    hist_months = selected_ticker.history(period='2y', interval='1mo')
    
    hist_days.dropna(axis=0, inplace=True)
    close_days = hist_days['Close'].reset_index()
    close_days = close_days.sort_values('Date' ,ascending=False)
    close_days.reset_index(drop=True, inplace=True)
    
    hist_weeks.dropna(axis=0, inplace=True)
    close_weeks = hist_weeks['Close'].reset_index()
    close_weeks = close_weeks.sort_values('Date' ,ascending=False)
    close_weeks.reset_index(drop=True, inplace=True)
    close_weeks = close_weeks.iloc[1:]
    close_weeks.reset_index(drop=True, inplace=True)
    
    hist_months.dropna(axis=0, inplace=True)
    close_months = hist_months['Close'].reset_index()
    close_months = close_months.sort_values('Date' ,ascending=False)
    close_months.reset_index(drop=True, inplace=True)
    close_months = close_months.iloc[1:]
    close_months.reset_index(drop=True, inplace=True)
    
    buy_nineth_turn_days = []
    buy_x_days = 0
    sell_nineth_turn_days = []
    sell_x_days = 0
    
    buy_nineth_turn_weeks = []
    buy_x_weeks = 0
    sell_nineth_turn_weeks = []
    sell_x_weeks = 0
    
    buy_nineth_turn_months = []
    buy_x_months = 0
    sell_nineth_turn_months = []
    sell_x_months = 0
    
    if len(close_days)>=14:
        for buy_day in range(10):
            buy_day_price = close_days.iloc[buy_day,1]
            buy_day_price_4prev = close_days.iloc[buy_day+4,1]
            if buy_day_price_4prev <= 12 and len(buy_nineth_turn_days) < 9:
                if buy_day_price < buy_day_price_4prev:
                    buy_x_days = buy_x_days + 1
                    buy_nineth_turn_days.append(buy_x_days)
                
                elif buy_day_price > buy_day_price_4prev:
                    buy_x_days = 0
                    buy_nineth_turn_days = []
                    
            elif len(buy_nineth_turn_days)==9 and buy_day_price > buy_day_price_4prev:
                buy_stocks_days.append(tickers)
    
        for sell_day in range(10):
            sell_day_price = close_days.iloc[sell_day,1]
            sell_day_price_4prev = close_days.iloc[sell_day+4,1]
            if buy_day_price_4prev <= 12 and len(sell_nineth_turn_days) < 9:
                if sell_day_price > sell_day_price_4prev:
                    sell_x_days = sell_x_days + 1
                    sell_nineth_turn_days.append(sell_x_days)
                
                elif sell_day_price < sell_day_price_4prev:
                    sell_x_days = 0
                    sell_nineth_turn_days = []
        
            elif len(sell_nineth_turn_days)==9 and sell_day_price < sell_day_price_4prev:
                sell_stocks_days.append(tickers)
                
    if len(close_weeks)>=14:
        for buy_week in range(10):
            buy_week_price = close_weeks.iloc[buy_week,1]
            buy_week_price_4prev = close_weeks.iloc[buy_week+4,1]
            if buy_week_price_4prev <= 12 and len(buy_nineth_turn_weeks) < 9:
                if buy_week_price < buy_week_price_4prev:
                    buy_x_weeks = buy_x_weeks + 1
                    buy_nineth_turn_weeks.append(buy_x_weeks)
                
                elif buy_week_price > buy_week_price_4prev:
                    buy_x_weeks = 0
                    buy_nineth_turn_weeks = []
                    
            elif len(buy_nineth_turn_weeks)==9 and buy_week_price > buy_week_price_4prev:
                buy_stocks_weeks.append(tickers)
    
        for sell_week in range(10):
            sell_week_price = close_weeks.iloc[sell_week,1]
            sell_week_price_4prev = close_weeks.iloc[sell_week+4,1]
            if buy_week_price_4prev <= 12 and len(sell_nineth_turn_weeks) < 9:
                if sell_week_price > sell_week_price_4prev:
                    sell_x_weeks = sell_x_weeks + 1
                    sell_nineth_turn_weeks.append(sell_x_weeks)
                
                elif sell_week_price < sell_week_price_4prev:
                    sell_x_weeks = 0
                    sell_nineth_turn_weeks = []
        
            elif len(sell_nineth_turn_weeks)==9 and sell_week_price < sell_week_price_4prev:
                sell_stocks_weeks.append(tickers)
                
    if len(close_months)>=14:
        for buy_month in range(10):
            buy_month_price = close_months.iloc[buy_month,1]
            buy_month_price_4prev = close_months.iloc[buy_month+4,1]
            if buy_month_price_4prev <= 12 and len(buy_nineth_turn_months) < 9:
                if buy_month_price < buy_month_price_4prev:
                    buy_x_months = buy_x_months + 1
                    buy_nineth_turn_months.append(buy_x_months)
                
                elif buy_month_price > buy_month_price_4prev:
                    buy_x_months = 0
                    buy_nineth_turn_months = []
                    
            elif len(buy_nineth_turn_months)==9 and buy_month_price > buy_month_price_4prev:
                buy_stocks_months.append(tickers)
    
        for sell_month in range(10):
            sell_month_price = close_months.iloc[sell_month,1]
            sell_month_price_4prev = close_months.iloc[sell_month+4,1]
            if buy_month_price_4prev <= 12 and len(sell_nineth_turn_months) < 9:
                if sell_month_price > sell_month_price_4prev:
                    sell_x_months = sell_x_months + 1
                    sell_nineth_turn_months.append(sell_x_months)
                
                elif sell_month_price < sell_month_price_4prev:
                    sell_x_months = 0
                    sell_nineth_turn_months = []
        
            elif len(sell_nineth_turn_months)==9 and sell_month_price < sell_month_price_4prev:
                sell_stocks_months.append(tickers)
        
print("TO BUY STOCKS LIST BY DAYS")
print(buy_stocks_days)
print("")
print("TO SELL STOCKS LIST BY DAYS")
print(sell_stocks_days)
print("")
print("TO BUY STOCKS LIST BY WEEKS")
print(buy_stocks_weeks)
print("")
print("TO SELL STOCKS LIST BY WEEKS")
print(sell_stocks_weeks)
print("")
print("TO BUY STOCKS LIST BY MONTHS")
print(buy_stocks_months)
print("")
print("TO SELL STOCKS LIST BY MONTHS")
print(sell_stocks_months)
```
